# Импорты

```python
import matplotlib.pyplot as plt

from dataset_creator.model.signals.signals import Signal, get_wave_fun, WaveFormType
from dataset_creator.model.signals.interval import Interval, stack_intervals, RangeAmplitudes, RangeFrequencies, ModulationType
from dataset_creator.model.signals.noise import WhiteNoise
```

# **1lin**
* Линейно-возрастающая от 20 Гц до 20 кГц синусоида с единичной амплитудой

```python
duration = 0.001
signal = Signal.build_signal(
  get_wave_fun(WaveFormType.sine),
  amplitudes = [
    Interval(duration=duration, values=RangeAmplitudes(1.0)),
  ],
  frequencies = [
    Interval(duration=duration, values=RangeFrequencies(20.0, 20_000.0, ModulationType.linear))
  ],
  k=1.0,
  dt=1e-6,
)
plt.plot(signal.t, signal.values)
plt.show()
```

# Пилообразный с шумом


```python
signal = Signal.build_signal(
    get_wave_fun(WaveFormType.sawtooth),
    amplitudes=[
        Interval(duration=2.0, values=RangeAmplitudes(1.0, 5.0, ModulationType.linear)),
        Interval(duration=2.0, values=RangeAmplitudes(5.0, 1.0, ModulationType.linear))
    ],
    frequencies=[Interval(duration=2.0, values=RangeFrequencies(1.0, 10.0)),
                 Interval(duration=2.0, values=RangeFrequencies(10.0, 1.0))],
    k=1.0,
    dt=0.00001,
    noise=WhiteNoise(level=-20, max_frequency=22000)
)
plt.plot(signal.t, signal.values)
plt.show()
```

# Гармонический (1, 3, 5 гармоники)

```python
signal = Signal.build_signal(
    get_wave_fun(WaveFormType.harmonic, harmonics=[1, 3, 5]),
    amplitudes=[
        Interval(duration=5.0, values=RangeAmplitudes(1.0, 5.0, ModulationType.logarithmic)),
    ],
    frequencies=[
        Interval(duration=5.0, values=RangeFrequencies(1.0, 3.0))
    ],
    k=1.0,
    dt=0.00001,
)
plt.plot(signal.t, signal.values)
plt.show()
```
